```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Read data

```python
df = pd.read_excel("data.xlsx", index_col="exp")
df.head()
```

```python
df.info()
```

### Extract data

```python
_X = df.iloc[:, :-3]
_Y = df.iloc[:, -3:]
print(_X.shape)
print(_Y.shape)
```

### Split data

```python
from sklearn.model_selection import train_test_split

_X_train, _X_test, _Y_train, _Y_test = train_test_split(
    _X, _Y, test_size=0.3, random_state=0
)

print(_X_train.shape)
print(_X_test.shape)
print(_Y_train.shape)
print(_Y_test.shape)
```

```python
_X_train
```

```python
from sklearn.preprocessing import StandardScaler

scX = StandardScaler()
scX.fit(_X_train)
X_train = scX.transform(_X_train)
X_test = scX.transform(_X_test)

scY = StandardScaler()
Y_train = scY.fit_transform(_Y_train)
Y_test = scY.transform(_Y_test)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
reg.fit(X_train, Y_train)
```

```python
Y_train_pred = reg.predict(X_train)
Y_test_pred = reg.predict(X_test)
print(Y_train_pred.shape)
print(Y_test_pred.shape)
```

```python
from datetime import datetime

dt = datetime.now().strftime("%Y-%m-%d_%H_%M")
```

```python
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_percentage_error,
    r2_score,
)

def eval_perf(y_true, y_pred):
    mse = mean_squared_error(y_true=y_true, y_pred=y_pred)
    mape = mean_absolute_percentage_error(y_true=y_true, y_pred=y_pred)
    r2 = r2_score(y_true=y_true, y_pred=y_pred)   
    return mse, mape, r2 

def print_perf(data):
    for k, v in data.items():
        print(k, " : ", v)
    
data_arr = []
for i in range(Y_train.shape[1]):
    mse_train, mape_train, r2_train = eval_perf(y_true=Y_train[:, i], y_pred=Y_train_pred[:, i])
    mse_test, mape_test, r2_test = eval_perf(y_true=Y_test[:, i], y_pred=Y_test_pred[:, i])
    
    data = {
        "Y": i + 1,
        "MSE Train": mse_train,
        "MSE Test": mse_test,
        "MAPE Train": mape_train,
        "MAPE Test": mape_test,
        "R2 Train": r2_train,
        "R2_Test": r2_test
    }
    data_arr.append(data)
    print_perf(data)

    
mse_train, mape_train, r2_train = eval_perf(y_true=Y_train, y_pred=Y_train_pred)
mse_test, mape_test, r2_test = eval_perf(y_true=Y_test, y_pred=Y_test_pred)
data = {
    "Y": "All",
    "MSE Train": mse_train,
    "MSE Test": mse_test,
    "MAPE Train": mape_train,
    "MAPE Test": mape_test,
    "R2 Train": r2_train,
    "R2_Test": r2_test
}
data_arr.append(data)
print_perf(data)

print(data_arr)

df_eval = pd.DataFrame.from_dict(data_arr)
```

```python
df_eval.to_excel(f"perf_{dt}.xlsx")
```

```python
from sklearn.metrics import PredictionErrorDisplay


for i in range(0, Y_train.shape[1]):
    fig, axes = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10,4), constrained_layout=True)
    display_train = PredictionErrorDisplay(y_true=Y_train[:, i], y_pred=Y_train_pred[:, i])
    display_train.plot(ax=axes[0])
    axes[0].set_title("Train")

    display_test = PredictionErrorDisplay(y_true=Y_test[:, i], y_pred=Y_test_pred[:, i])
    display_test.plot(ax=axes[1])
    axes[1].set_title("Test")
    
    fig.savefig(f"file-{dt}-{i+1}.png", dpi=300)
    plt.show()
```

```python
data_save = {
    "model": reg,
    "desc": "This is linear regression run Ex1",
    "Y_train": Y_train,
    "Y_train_pred": Y_train_pred,
    "Y_test": Y_test,
    "Y_test_pred": Y_test_pred,
    "X_train": X_train,
    "X_test": X_test,
    "eval": df_eval
}
```
